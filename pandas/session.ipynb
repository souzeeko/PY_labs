```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('session.csv', sep =';')
df.head()
```

```python
df['Programming'].value_counts().plot(kind = 'bar', title = 'Programming mark statistics', xlabel = 'Mark', ylabel = 'Students amount')
```

```python
df['English'].value_counts().plot(kind = 'bar', title = 'English mark statistics', xlabel = 'Mark', ylabel = 'Students amount')
```

```python
df['Mathematics'].value_counts().plot(kind = 'bar', title = 'Mathematics mark statistics', xlabel = 'Mark', ylabel = 'Students amount')
```

```python
df['Project'].value_counts().plot(kind = 'bar', title = 'Project mark statistics', xlabel = 'Mark', ylabel = 'Students amount')
```

```python
df.pivot_table(values = ['Programming', 'English', 'Mathematics', 'Project'], index = 'Sex', aggfunc = 'mean').plot(kind = 'bar', title = 'Avg. marks to subjects diogram', ylabel = 'Avg. mark')
```
